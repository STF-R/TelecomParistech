### Question 4 

```python
def perceptronCyclique(X, y, grad, eps=0.01, w=[0,0,0], niter = 500):
    j=0
    while j < niter :
        for i in range(y.size):
            wnew = w - eps*grad(X,y,w)
        w = wnew
        j+=1
    return w
```
