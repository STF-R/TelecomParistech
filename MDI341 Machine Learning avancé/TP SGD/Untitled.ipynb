### 2-3-4) implémentation de SGD

```python
def algo_SGD(X, y, nb_iter):
    def gamma(t):
        return 1. / (100 + t)
    t = 0
    n_samples = X.shape[0]
    w0 = np.zeros(X.shape[1])
    while t<nb_iter:
        i = random.randint(0, n_samples-1)
        w1 = w0 -  gamma(t) * X[i,:] * (np.dot(X[i], w0)-y[i])
        t+=1
        w0 = w1
    return w1

n_samples = 1000
n_features = 100
n_iter = 1000  # number of iterations
gamma = 0.01  # step size

X_toy = np.random.randn(n_samples, n_features)
w_target = np.ones(n_features)
y_toy = X_toy.dot(w_target)

p = []
for i in range(10,1500,20):
    res = algo_SGD(X_toy, y_toy, nb_iter=i)
    p.append(0.5 * np.mean((res - w_target) ** 2))
    
plt.figure()
plt.plot(range(10,1500,20), p)
plt.xlabel("Nombre itérations")
plt.ylabel("loss")
plt.title("loss in fuction of the iterations")
plt.show()
```
